```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ssl_df = pd.read_csv('data/Strategic_Subject_List.csv')
```

```python
plt.rcParams["figure.figsize"] = [15, 5]

ssl_df.hist('SSL_SCORE')
```

```python
plt.rcParams["figure.figsize"] = [17, 6]

white_ssl_df = ssl_df.loc[ssl_df['RACE_CODE_CD'] == 'WHI'].copy()
black_ssl_df = ssl_df.loc[ssl_df['RACE_CODE_CD'] == 'BLK'].copy()
hisp_ssl_df = ssl_df.loc[ssl_df['RACE_CODE_CD'] == 'WWH'].copy()

fig, axs = plt.subplots(1, 3)
print('SSL Score Distributions: White, Black, Hispanic')
white_ssl_df.hist('SSL_SCORE', ax=axs[0])
hisp_ssl_df.hist('SSL_SCORE', ax=axs[1])
black_ssl_df.hist('SSL_SCORE', ax=axs[2])
```

```python
lt_30_ssl_df = ssl_df.loc[ssl_df['AGE_GROUP'].isin(['less than 20', '20-30'])].copy()
gte_30_ssl_df = ssl_df.loc[~ssl_df['AGE_GROUP'].isin(['less than 20', '20-30'])].copy()

fig, axs = plt.subplots(1, 2)
print('SSL Score Distributions: Age < 30, Age >= 30')
lt_30_ssl_df.hist('SSL_SCORE', ax=axs[0])
gte_30_ssl_df.hist('SSL_SCORE', ax=axs[1])
```

```python
m_ssl_df = ssl_df.loc[ssl_df['SEX_CODE_CD'] == 'M'].copy()
f_ssl_df = ssl_df.loc[ssl_df['SEX_CODE_CD'] == 'F'].copy()

fig, axs = plt.subplots(1, 2)
print('SSL Score Distributions: Male, Female')
m_ssl_df.hist('SSL_SCORE', ax=axs[0])
f_ssl_df.hist('SSL_SCORE', ax=axs[1])
```

```python
ybm_ssl_df = ssl_df.loc[(ssl_df['SEX_CODE_CD'] == 'M') &
                        (ssl_df['RACE_CODE_CD'] == 'BLK') &
                        (ssl_df['AGE_GROUP'].isin(['less than 20', '20-30']))].copy()
ywm_ssl_df = ssl_df.loc[(ssl_df['SEX_CODE_CD'] == 'M') &
                        (ssl_df['RACE_CODE_CD'] == 'WHI') &
                        (ssl_df['AGE_GROUP'].isin(['less than 20', '20-30']))].copy()

fig, axs = plt.subplots(1, 2)
print('SSL Score Distributions (Men 18-29): Black, White')
ybm_ssl_df.hist('SSL_SCORE', ax=axs[0])
ywm_ssl_df.hist('SSL_SCORE', ax=axs[1])
```
